```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd 
```

```python
url = 'https://en.wikipedia.org/wiki/Forbes_list_of_the_most_valuable_football_clubs'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
table = soup.find('table')
print(table)
```

```python
headers = table.find_all('th')
```

```python
print(headers)
```

```python
columns = [header.text.strip() for header in headers]
columns
```

```python
df = pd.DataFrame(columns = columns)
df
```

```python
pre_rows = table.find_all('tr')
```

```python
for row in pre_rows[1:]:
    row_data = row.find_all('td')
    rows = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = rows
    
```

```python
df
```

```python
df.to_csv('richestclubs.csv', index = False)
```
